```python
def preprocess_data(data_frame):
    X = data_frame.replace('unknown', np.nan)
    num_columns = X.select_dtypes(exclude='object').columns
    
    
    cat_job = CategoricalDtype(categories=[
        'blue-collar', 
        'entrepreneur', 
        'services', 
        'unemployed', 
        'technician', 
        'admin.', 
        'retired', 
        'housemaid', 
        'student', 
        'self-employed', 
        'management'
    ])

    cat_marital = CategoricalDtype(categories=[
        'single',
        'married',
        'divorced'
    ])

    cat_education = CategoricalDtype(
        categories=[
            'illiterate',
            'basic.4y',
            'basic.6y',
            'basic.9y',
            'professional.course',
            'high.school',
            'university.degree'
        ],
        ordered=True
    )

    cat_default = CategoricalDtype(categories=['yes', 'no'])
    cat_housing = CategoricalDtype(categories=['yes', 'no'])
    cat_loan = CategoricalDtype(categories=['yes', 'no'])
    cat_contact = CategoricalDtype(categories=['telephone', 'cellular'])
    cat_poutcome = CategoricalDtype(categories=['failure', 'nonexistent', 'success'])

    X.job = X.job.astype(cat_job)
    X.marital = X.marital.astype(cat_marital)
    X.education = X.education.astype(cat_education)

    X.default = X.default.astype(cat_default)
    X.housing = X.housing.astype(cat_housing)
    X.loan = X.loan.astype(cat_loan)
    X.contact = X.contact.astype(cat_contact)
    X.poutcome = X.poutcome.astype(cat_poutcome)

    cat_columns = X.select_dtypes(['category']).columns
    X[cat_columns] = X[cat_columns].apply(lambda x: x.cat.codes)
    
    month_numbers = list(range(1, 13))
    months_abbr = [ calendar.month_abbr[idx] for idx in month_numbers ]

    months_dict = dict(zip(months_abbr, month_numbers))
    X.month = X.month.apply(str.capitalize)
    X.month = X.month.map(months_dict)
    
    day_numbers = list(range(7))
    days_abbr = [ calendar.day_abbr[idx] for idx in day_numbers ]
    days_dict = dict(zip(days_abbr, day_numbers))

    X.day_of_week = X.day_of_week.apply(str.capitalize)
    X.day_of_week = X.day_of_week.map(days_dict)
        
    return X
```

```python
def imput_missing_data(data):
    X_missing = data.as_matrix()
    
    imp = Imputer(missing_values=-1, strategy='most_frequent')
    
    X_imputed = imp.fit_transform(X_missing)
    
    return pd.DataFrame(data=X_imputed, columns=data.columns, index=data.index)
    
```

```python
new_train = new_train.drop(['default', 'housing', 'loan'], axis=1)
```

```python
new_train.columns
```

```python
y = list(new_train['target'])
new_train.drop('target', axis=1, inplace=True)
```

```python
prep_test = preprocess_data(test) 
new_test = imput_missing_data(prep_test)

new_test = new_test.drop(['default', 'housing', 'loan'], axis=1)
```

```python
num_columns = ['age',
    'duration', 'campaign', 'pdays', 'previous', 'emp.var.rate',
    'cons.price.idx', 'cons.conf.idx', 'euribor3m', 'nr.employed'
]

all_data = new_train.append(new_test)
scaler = StandardScaler().fit(all_data[num_columns])

new_train[num_columns] = scaler.transform(new_train[num_columns])
new_test[num_columns] = scaler.transform(new_test[num_columns])
```

```python
X_train, X_test, y_train, y_test = train_test_split(new_train, y, test_size=0.33, random_state=42, shuffle=True)
```

```python
X_train.shape
```

```python
decision_tree_classifier = DecisionTreeClassifier()

parameter_grid = {'max_depth': [3, 5, 10, 15, 20],
                  'max_features': [2, 5, 10, 17]}

cross_validation = StratifiedKFold(n_splits=10)

grid_search = GridSearchCV(decision_tree_classifier, param_grid = parameter_grid,
                          cv = cross_validation, scoring='roc_auc', n_jobs=-1)

grid_search.fit(X_train, y_train)
```

```python
print("Best DecisionTreeClassifier Score: {}".format(grid_search.best_score_))
print("Best DecisionTreeClassifier params: {}".format(grid_search.best_params_))
print("Best DecisionTreeClassifier estimator: {}".format(grid_search.best_estimator_))
```

```python
param_grid = {'solver': ['newton-cg', 'sag', 'lbfgs']}

gs_lr = GridSearchCV(LogisticRegression(penalty='l2', class_weight='balanced'), param_grid, cv=cross_validation, 
                     scoring='roc_auc', n_jobs=-1)

gs_lr.fit(X_train, y_train)
```

```python
print("Tuned Logistic Regression Parameters: {}".format(gs_lr.best_params_)) 
print("Best score is {}".format(gs_lr.best_score_))
print("Best Logistic Regression estimator is: {}".format(gs_lr.best_estimator_))
```

```python
k_range = list(range(9, 21)) 
weight_options = ['uniform', 'distance']
param_grid = dict(n_neighbors=k_range, weights=weight_options)
knn = KNeighborsClassifier()

grid_knn = GridSearchCV(knn, param_grid, cv=cross_validation, scoring='roc_auc', n_jobs=-1)
grid_knn.fit(X_train, y_train)
```

```python
print("Best KNN params: {}".format(grid_knn.best_params_))
print("Best KNN Score: {}".format(grid_knn.best_score_))
print("Best KNN estimator: {}".format(grid_knn.best_estimator_))
```

```python
plt.figure(figsize=(10, 7))

fpr_t, tpr_t, thresholds_t = roc_curve(y_test, predict_proba_lr_tuned[:,1])
roc_auc_t  = auc(fpr_t, tpr_t)
plt.plot(fpr_t, tpr_t, label='%s ROC (area = %0.4f)' % ('LR tuned', roc_auc_t))

fpr_dtc, tpr_dtc, thresholds_dtc = roc_curve(y_test, predict_proba_dtc_tuned[:,1])
roc_auc_dtc  = auc(fpr_dtc, tpr_dtc)
plt.plot(fpr_dtc, tpr_dtc, label='%s ROC (area = %0.4f)' % ('DTC tuned', roc_auc_dtc))

fpr_knn, tpr_knn, thresholds_knn = roc_curve(y_test, predict_proba_knn_tuned[:,1])
roc_auc_knn  = auc(fpr_knn, tpr_knn)
plt.plot(fpr_knn, tpr_knn, label='%s ROC (area = %0.4f)' % ('kNN tuned', roc_auc_knn))

fpr_rfc, tpr_rfc, thresholds_rfc = roc_curve(y_test, rfc_pb[:,1])
roc_auc_rfc  = auc(fpr_rfc, tpr_rfc)
plt.plot(fpr_rfc, tpr_rfc, label='%s ROC (area = %0.4f)' % ('RFC tuned', roc_auc_rfc))

plt.plot([0, 1], [0, 1])
plt.ylabel('tpr')
plt.xlabel('fpr')
plt.grid(True)
plt.title('ROC curve')
plt.xlim((-0.01, 1.01))
plt.ylim((-0.01, 1.01))
plt.legend(loc=0, fontsize='small')
```
